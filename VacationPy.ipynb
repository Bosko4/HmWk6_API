### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.

```python
hotel_df['Hotel Name'] = ""
hotel_df['Hotel City'] = ""
hotel_df['Hotel Country'] = ""
hotel_df['Map Code'] = ""
hotel_df['Location'] = ""

hotel_df.head()
```

```python
#---Google place

params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key
}

#for i in range(len(hotel_df)):
for index, row in hotel_df.iterrows():
    lat = row['Latitude']
    lon = row['Longitude']

   
    params["location"] = f"{lat},{lon}"

    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"    
    response = requests.get(base_url, params=params)    
    #print(response.url) #--- used to evalutate code was properly generating query URLs.
    response = response.json()
    
    try:
        hotel_df.loc[index, "Hotel Name"] = response["results"][0]["name"]
        hotel_df.loc[index, "Location"] = response["results"][0]["plus_code"]["compound_code"]
        hotel_df.loc[index, "Map Code"] = response["results"][0]["plus_code"]["global_code"]     
        
    except (KeyError, IndexError):
        print("No Results")
        
```

```python
#---break out city/country of closest hotel results
for index, row in hotel_df.iterrows():
    loc = row['Location'][8:]
    loc1 = loc.split(",")
    
    hotel_df.loc[index, "Hotel City"] = loc1[0]
    hotel_df.loc[index, "Hotel Country"] = loc1[-1]
    #print(loc1)
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{Hotel City}</dd>
<dt>Country</dt><dd>{Hotel Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Latitude", "Longitude"]]
```

```python
# Add marker layer ontop of heat map
fig = gmaps.figure()

hotels = gmaps.marker_layer(locations,info_box_content=hotel_info)

# Display figure
fig.add_layer(hotels)
fig
```
